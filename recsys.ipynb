```python
#---IMPORTAÇÃO DE BIBLIOTECAS USADAS NO PROJETO---#
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#---OBTENDO OS DADOS UTILIZADOS, NO FORMATO .CSV E TRANSFORMANDO EM OBJETOS DATAFRAMES COM A BIBLIOTECA PANDAS---#
filmes = pd.read_csv('C:/Users/João Pedro Mendes/Documents/JupyterMeuLab/Lab XI - CPBR/Data/movies.csv')
avaliacoes = pd.read_csv('C:/Users/João Pedro Mendes/Documents/JupyterMeuLab/Lab XI - CPBR/Data/ratings.csv')
#---DUAS VARIÁVEIS, UMA COM OS FILMES E SUAS INFORMAÇÕES E OUTRA COM AS AVALIAÇÕES E SUAS INFORMAÇÕES---#
```

```python
#---JUNTANDO OS DATAFRAMES ANTERIORES EM UM ÚNICO DATAFRAME E VISUALIZANDO PARA ENTENDER SUA ESTRUTURA---#
dfinfo = pd.merge(filmes, avaliacoes)

#---VERIFICANDO O DATAFRAME---#
dfinfo
```

```python
#---REMOÇÃO DE COLUNAS QUE NÃO SERÃO UTILIZADAS DURANTE O PROJETO---#
dfinfo.drop(columns={'timestamp', 'genres'}, inplace=True)

#---VERIFICANDO AS 3 PRIMEIRAS LINHAS DO DATAFRAME---#
dfinfo.head(3)
```

```python
#---CRIANDO UM NOVO DATAFRAME PARA GERAR NOSSOS DADOS EM FORMATO DE TABELA DINÂMICA, FACILITANDO A OBSERVAÇÃO DO DF---#
avaliacao_usuario = dfinfo.pivot_table(index=['userId'], columns=['title'], values='rating')

#---VERIFICANDO O DATAFRAME---#
avaliacao_usuario
```

```python
#---ELIMINAMOS OS FILME COM MENOS DE 10 AVALIAÇÕES E PREENCHEMOS OS DADOS AUSENTES POR 0---#
avaliacao_usuario = avaliacao_usuario.dropna(thresh=3, axis=1).fillna(0)

#---VERIFICANDO O DATAFRAME---#
avaliacao_usuario
#---OBSERVAMOS MUITOS VALORES 0.0, QUE ANTERIORMENTE ERAM NaN TENDO EM VISTA QUE NEM TODOS OS USUÁRIOS VIRAM AQUELE DETERMINADO FILME---#
```

```python
#---CRIANDO UMA FUNÇÃO QUE PADRONIZA AS AVALIAÇÕES EM VALORES ENTRE -1 E 1---#
def padronizar(linha):
    
    nova_linha = (linha - linha.mean()/(linha.max()-linha.min()))
    #---(X-MÉDIA(X))/(MÁXIMO(X)-MÍNIMO(X))---#             
    return nova_linha
    #---RETORNAMOS NOSSA NOVA LINHA COMO O RESULTADO DA FUNÇÃO---#

novaavaliacao = avaliacao_usuario.apply(padronizar)
#---FORMAMOS UM NOVO DATAFRAME A PARTIR DA APLICAÇÃO DA NOSSA FUNÇÃO PADRONIZAR NO NOSSO DATAFRAME COM AS AVALIAÇÕES DE USUÁRIOS---#

similaridade_filmes = cosine_similarity(novaavaliacao.T)
#---UTILIZANDO O RECURSO COSINE_SIMILARITY, DA BIBLIOTECA SKLEARN, CRIAMOS UM ARRAY QUE OBTEM OS VALORES DA VARIAVEL---#
#---NOVAAVALIACAO E REALIZA A FUNÇÃO BUILT-IN COSINE_SIMILARITY(SIMILARIDADE PELO COSSENO)---#
print(similaridade_filmes)
#---VERIFICANDO NOSSO ARRAY---#
```

```python
#---NOVO DATAFRAME COM A VARIAVEL SIMARILIDADE_FILMES, UTILIZANDO COMO ÍNDICE E COMO COLUNAS, AS COLUNAS---#
#---DO NOSSO DF AVALIACAO_USUARIO---#
dfsimilaridade = pd.DataFrame(similaridade_filmes, index=avaliacao_usuario.columns, columns=avaliacao_usuario.columns)

#---VERIFICANDO O DATAFRAME---#
dfsimilaridade
```

```python
#---FUNÇÃO QUE UTILIZA COMO PARÂMETROS, O NOME DO FILME E A NOTA DO USUÁRIO---#
def nota_similaridade(nome_do_filme, nota_do_usuario):
    nota = dfsimilaridade[nome_do_filme]*(nota_do_usuario-2.5)
    #---BUSCAMOS DENTRO DO NOSSO DATAFRAME DE SIMILARIDADE, O NOME DO FILME E MULTIPLICAMOS PELA NOTA DO USUÁRIO - 2.5---#
    #---2.5 É A METADE DA NOTA MÁXIMA, 5, A SUBTRAÇÃO POR 2.5 DEIXA OS VALORES MAIS COMPREENSÍVEIS NA HORA DE DEFINIR---#
    #---SE AQUELE FILME É BOM OU NÃO PARA O USUÁRIO---#
    nota = nota.sort_values(ascending=False)
    #---ANTES DE FINALIZAR A FUNÇÃO, RETORNAMOS OS VALORES DA VARIÁVEL NOTA EM ORDEM DECRESCENTE---#
    
    return nota
    #---RETORNAMOS A VARIÁVEL NOTA COMO RESULTADO DA FUNÇÃO---#
```

```python
#---NA LISTA MEUUSUARIO, COLOCAMOS OS FILMES, SEU ANO DE LANÇAMENTO ENTRE PARENTESES E APÓS ISSO, A NOTA---#
meuusuario = [("Shutter Island (2010)", 4),("Twilight (2008)", 5), ("Lemony Snicket's A Series of Unfortunate Events (2004)", 5)] 
              #("Chinatown (1974)", 5)]

filmes_similares = pd.DataFrame()
#---CRIAMOS OUTRO DF, PARA GUARDAR OS TÍTULOS DOS FILMES E O ÍNDICE DE SIMILARIDADE---#
for filme, nota_user in meuusuario:
    #---NO FOR LOOP, USAMOS DOIS ITENS, FILMES E A NOTA DO USUÁRIO DENTRO DA NOSSA LISTA MEUUSUARIO---#
    filmes_similares = filmes_similares.append(nota_similaridade(filme, nota_user), ignore_index=True)
    #---UTILIZAMOS A FUNÇÃO NOTA_SIMILARIDADE E ADICIONAMOS OS RESULTADOS NO NOSSO DF---#
filmes_similares.sum().sort_values(ascending=False)
#---FAZEMOS A SOMA DOS VALORES DE SIMILARIDADE E VISUALIZAMOS OS 50 COM MAIOR PROBABILIDADE DO USUARIO GOSTAR---#
```
